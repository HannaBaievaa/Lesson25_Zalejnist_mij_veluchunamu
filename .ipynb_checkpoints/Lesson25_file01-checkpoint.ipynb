### Завдання 1

Ймовірність того, що при перевірці деталь виявиться стандартною дорівнює 0,6. Перевірці підлягають 3 деталі. Побудувати закон розподілу системи двох дискретних величин:
- $X$ --- поява числа бракованих деталей;
- $Y$ --- поява числа стандартних деталей.

Обчислити коефіцієнт кореляції $r_{XY}$

#### Рішення
Якщо ймовірність появи стандартної деталі 0,6, тоді поява бракованої буде дорівнювати $1-0,6=0,4$

У задачі можливі чотири випадки:
1. Всі деталі є бракованими
2. Одна бракована, дві стандартні (три варіанти)
3. Дві браковані, одна стандартна (три варіанти)
4. Всі деталі браковані

Необхідно розрахувати ймовірності цих подій:

- $P(X = 0, Y = 3) = 0.6 \times 0.6 \times 0.6 = 0.216$
- $P(X = 1, Y = 2) = 0.4 \times 0.6 \times 0.6 \times 3 = 0.432$
- $P(X = 2, Y = 1) = 0.4 \times 0.4 \times 0.6 \times 3 = 0.288$
- $P(X = 3, Y = 0) = 0.4 \times 0.4 \times 0.4 = 0.064$

Побудова закону розподілу випадкових величин:

| **X** |   0   |   1   |   2   |   3   |
|:-----:|:-----:|:-----:|:-----:|:-----:|
| **P** | 0.216 | 0.432 | 0.288 | 0.064 |

| **Y** |   0   |   1   |   2   |   3   |
|:-----:|:-----:|:-----:|:-----:|:-----:|
| **P** | 0.064 | 0.288 | 0.432 | 0.216 |

Математичне сподівання кожної величини:
$M_X = 0 \times 0.216 + 1 \times 0.432 + 2 \times 0.288 + 3 \times 0.064 = 1.2$

$M_Y = 0 \times 0.064 + 1 \times 0.288 + 2 \times 0.432 + 3 \times 0.216 = 1.8$

Дисперсія:
$D_X = 0^2 \times 0.216 + 1^2 \times 0.432 + 2^2 \times 0.288 + 3^2 \times 0.064 - 1.2^2 = 0.72$

$D_Y = 0^2 \times 0.064 + 1^2 \times 0.288 + 2^2 \times 0.432 + 3^2 \times 0.216 - 1.8^2 = 0.72$

Коваріація:
$cov(X/Y) = 1 \times 2 \times 0.432 + 2 \times 1 \times 0.288 - 1.2 \times 1.8 = -0.72$

Кореляція:
$r_{XY} = \frac{-0.72}{\sqrt{0.72} \times \sqrt{0.72}} = -1$

**Відповідь**: оскільки $r_{XY}=-1$, тоді у двох випадкових величин спостерігається зворотній взаємозв'язок.
